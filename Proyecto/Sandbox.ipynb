```python
import networkx as nx
```

```python
G_Trump = nx.read_edgelist('./data/trump_graph.csv',create_using=nx.Graph(),nodetype=int)
```

```python
deg_Trump = G_Trump.degree
print(G_Trump.degree)
```

```python
deg_min = 15
nodes = [k[0] for k in deg_Trump if k[1]>=deg_min]
print(len(nodes))
G_simple = G_Trump.subgraph(nodes)
```

```python
print(nx.info(G_simple))
```

```python
cc = nx.connected_components(G_simple)
components_size = [len(c) for c in sorted(cc, key=len, reverse=True)]
```

```python
components_size
```

```python
max(components_size)
```

```python
largest_cc = max(nx.connected_components(G_simple), key=len)
```

```python
type(largest_cc)
```

```python
G_final = G_simple.subgraph(largest_cc)
```

```python
print(nx.info(G_final))
print(f'Densidad: {nx.density(G_final)}')
diameter = nx.diameter(G_final) #Found infinite path length because the graph is not connected
print(f'diametro: {diameter}') #todos los nodos estn conectados y ahora tiene diametro
triadic_closure = nx.transitivity(G_final)
print(f'Transitividad: {triadic_closure}')
```

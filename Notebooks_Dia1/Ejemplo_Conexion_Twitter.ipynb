```python
import tweepy
from twitter_secrets import CONSUMER_KEY, CONSUMER_SECRET, ACCESS_TOKEN, ACCESS_TOKEN_SECRET

auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)

api = tweepy.API(auth)

#timeline de el due√±o de acces_token
#Returns the 20 most recent statuses, including retweets, posted by the authenticating user and that user‚Äôs friends. This is the equivalent of /timeline/home on the Web.
public_tweets = api.home_timeline()

print(type(public_tweets))
print("-------")

for tweet in public_tweets:
    print(tweet.text)
    print("-------")
```
